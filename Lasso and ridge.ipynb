```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score

# Load the California Housing Dataset
data = fetch_california_housing()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target

# Standardize the dataset
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
lin_coefficients = lin_reg.coef_
lin_r2 = r2_score(y_test, lin_reg.predict(X_test))

# User-defined alpha values
ridge_alpha = float(input("Enter alpha value for Ridge Regression: "))
lasso_alpha = float(input("Enter alpha value for Lasso Regression: "))

# Ridge Regression (L2 Regularization)
ridge_reg = Ridge(alpha=ridge_alpha)
ridge_reg.fit(X_train, y_train)
ridge_coefficients = ridge_reg.coef_
ridge_r2 = r2_score(y_test, ridge_reg.predict(X_test))

# Lasso Regression (L1 Regularization)
lasso_reg = Lasso(alpha=lasso_alpha)
lasso_reg.fit(X_train, y_train)
lasso_coefficients = lasso_reg.coef_
lasso_r2 = r2_score(y_test, lasso_reg.predict(X_test))

# Plot the coefficients
features = data.feature_names
x = np.arange(len(features))
width = 0.25

fig, ax = plt.subplots(figsize=(15, 8))
ax.bar(x - width, lin_coefficients, width, label='Linear Regression')
ax.bar(x, ridge_coefficients, width, label='Ridge Regression')
ax.bar(x + width, lasso_coefficients, width, label='Lasso Regression')

ax.set_xlabel('Features')
ax.set_ylabel('Coefficient Value')
ax.set_title('Comparison of Coefficients in Linear, Ridge, and Lasso Regression')
ax.set_xticks(x)
ax.set_xticklabels(features, rotation=45)
ax.legend()

plt.tight_layout()
plt.show()

# Display Coefficients and R2 Scores
coefficients_df = pd.DataFrame({
    'Feature': features,
    'Linear': lin_coefficients,
    'Ridge': ridge_coefficients,
    'Lasso': lasso_coefficients
})

print(coefficients_df)
print(f"R2 Score (Linear Regression): {lin_r2:.4f}")
print(f"R2 Score (Ridge Regression, alpha={ridge_alpha}): {ridge_r2:.4f}")
print(f"R2 Score (Lasso Regression, alpha={lasso_alpha}): {lasso_r2:.4f}")
```

From the above results, we can observe that Ridge Regression (with α = 0.01) has made minimal changes to the coefficients compared to Linear Regression, indicating that Ridge primarily prevents overfitting by slightly shrinking the coefficients without eliminating any features.

On the other hand, Lasso Regression (with α = 0.01) has not only increased the R² score but also reduced the magnitude of several coefficients significantly. Notably, the coefficient for the feature 'Population' was reduced to zero, suggesting that this feature does not contribute significantly to the model's predictions. This sparsity property of Lasso can be useful for feature selection, as it helps identify and remove irrelevant features, simplifying the model while maintaining or even improving predictive performance.

# **Comparison of Linear, Ridge, and Lasso Regression Across Different Alpha Values**

- **Linear Regression:** No regularization; coefficients are directly optimized to minimize loss.  
- **Ridge Regression:** Shrinks coefficients proportionally but does **not eliminate** features (no coefficients become zero).  
- **Lasso Regression:** Shrinks coefficients and can **eliminate features** by setting some coefficients to **zero**.

- As **α (alpha)** increases:
   - **Ridge Regression:** Coefficient shrinkage becomes slightly more pronounced, but no coefficients are set to zero.

   - **R² Score:** Ridge maintains stability, while Lasso shows fluctuations based on the level of regularization.

# **Understanding the Benefits of Lasso and Ridge Regularization**

### **1. Overview of Regularization:**
Regularization techniques like **Lasso** and **Ridge** introduce a penalty term to the loss function to prevent overfitting and improve generalization. They help models avoid overly complex solutions by controlling the magnitude of coefficients or weights.

---

### **2. Applications in Regression:**

#### **a. Lasso Regression:**
- **Key Feature:** Adds the absolute value of coefficients as a penalty (\( \lambda \sum |w_i| \)).
- **Benefits:**
  1. **Feature Selection:** Drives some coefficients to **exactly zero**, effectively eliminating less important features.
  2. **Model Simplification:** Reduces dimensionality, making the model more interpretable.
  3. **Improved Generalization:** Prevents overfitting by focusing on the most impactful features.

#### **b. Ridge Regression:**
- **Key Feature:** Adds the squared value of coefficients as a penalty (\( \lambda \sum w_i^2 \)).
- **Benefits:**
  1. **Weight Shrinkage:** Reduces the magnitude of coefficients, preventing any single feature from dominating the predictions.
  2. **Robustness:** Handles multicollinearity (correlated features) better by balancing contributions across all features.
  3. **No Feature Elimination:** Ensures all features remain in the model, which is useful when all variables are considered relevant.

---

### **3. Applications in Neural Networks:**

#### **a. Lasso Regularization:**
- **Key Feature:** Adds the absolute sum of the weights to the loss function.
- **Benefits:**
  1. **Sparse Connections:** Encourages some weights to become **exactly zero**, leading to **sparser neural networks**.
  2. **Feature Selection:** Helps identify which inputs (features) are most important for the network's predictions.
  3. **Simpler Models:** Reduces the complexity of the network, making it faster and easier to interpret.

#### **b. Ridge Regularization (Weight Decay):**
- **Key Feature:** Adds the squared sum of weights to the loss function.
- **Benefits:**
  1. **Weight Stabilization:** Reduces the magnitude of weights uniformly, avoiding over-reliance on any particular neuron or feature.
  2. **Improved Generalization:** Helps prevent overfitting by controlling the growth of weights, especially in deep networks.
  3. **Smooth Training:** Ensures smoother gradients during optimization, leading to more stable convergence.

---

### **4. Key Differences Between Lasso and Ridge:**

| **Aspect**               | **Lasso (L1)**                          | **Ridge (L2)**                          |
|---------------------------|-----------------------------------------|-----------------------------------------|
| **Penalty Term**          | \( \lambda \sum |w_i| \)               | \( \lambda \sum w_i^2 \)               |
| **Effect on Coefficients**| Drives some weights to zero            | Shrinks weights uniformly               |
| **Feature Selection**     | Yes                                    | No                                      |
| **Sparsity**              | Induces sparsity                       | Does not induce sparsity                |
| **Best For**              | Sparse data, feature selection         | Balanced weight shrinkage, all features relevant |

---

### **5. Combined Approach: Elastic Net**
- **Feature:** Combines L1 (Lasso) and L2 (Ridge) regularization into one penalty term:
  \[
  \text{Loss} = \text{Original Loss} + \lambda_1 \sum |w_i| + \lambda_2 \sum w_i^2
  \]
- **Benefits:**
  1. Combines the **sparsity** of Lasso with the **stability** of Ridge.
  2. Useful for high-dimensional datasets with correlated features.

---

### **6. Summary of Use Cases:**

| **Model Type**            | **Lasso (L1)**                          | **Ridge (L2)**                          |
|---------------------------|-----------------------------------------|-----------------------------------------|
| **Linear Regression**     | Feature selection, sparse coefficients | Robustness to multicollinearity         |
| **Neural Networks**       | Sparse connections, feature selection  | Weight stabilization, improved generalization |

---

### **7. Key Takeaways:**
- **Lasso (L1)** is best when feature selection and sparsity are required.
- **Ridge (L2)** is ideal for robust models that use all features while preventing overfitting.
- In **neural networks**, these regularization techniques act as constraints on weights, leading to better generalization and smoother training.
